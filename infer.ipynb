```python
import torch
from utils import generate, ids_to_text, text_to_ids
from dataset import format_for_generate
from model import Model
from config import cfg
```

```python
model = Model(cfg=cfg)
```

```python
state = torch.load("weights_only_10.pt", map_location="cpu")
model.load_state_dict(state["model_state_dict"])
```

```python
import tiktoken
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
device = "cpu"
```

```python
ids = text_to_ids(text=format_for_generate(text="who created you?"), tokenizer=tokenizer)
print(ids_to_text(ids=ids, tokenizer=tokenizer))
generated_text = generate(cfg=cfg, device=device, eos=50256, max_new_tokens=cfg["max_new_tokens"], model=model, ids=ids, tokenizer=tokenizer)

for token in generated_text:
    print(token, end="", flush=True)
```
